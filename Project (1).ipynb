```python
df['date'] = pd.to_datetime(df['date'])
```

```python
selected_countries = ['Kenya', 'United States', 'India']
plt.figure(figsize=(10, 6))
for country in selected_countries:
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], country_data['total_cases'], label=country)
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.title('Total Cases Over Time for Selected Countries')
plt.legend()
plt.show()
```

```python
country_data = ['Kenya', 'United States', 'India']
plt.figure(figsize=(12,6))
for country in country_data:
  country_data = df[df['location'] == country].sort_values('date')
  plt.plot(country_data['date'], country_data['total_deaths'], label= country)
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.title('Total Deaths Over Time')
plt.legend()
plt.show()
```

```python
df['Death Rate'] = df['total_deaths'] / df['total_cases'] * 100
print(df[['location', 'total_deaths', 'total_cases', 'Death Rate']])
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['date'], df['total_cases'], label='Total Cases', color = 'blue')
plt.plot(df['date'], df['total_deaths'], label='Total Deaths', color = 'red')
```

```python
latest_cases = df.groupby('location')['total_cases'].max().sort_values(ascending=False)
top_n = 10
top_countries = latest_cases.head(top_n)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='viridis')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.title(f'Top {top_n} Countries with the Highest Total Cases')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
numerical_cols = ['total_cases', 'total_deaths', 'new_cases', 'Death Rate']
df_corr = df[numerical_cols]
correlation_matrix = df_corr.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Covid-19 Data')
plt.tight_layout()
plt.show()
```

```python
selected_countries = ['Kenya', 'United States', 'India']
plt.figure(figsize=(12, 6))
for country in selected_countries:
    country_data = df[df['location'] == country].sort_values('date')
    plt.plot(country_data['date'], country_data['total_vaccinations'], label=country)
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.title('Total Vaccinations Over Time for Selected Countries')
plt.legend()
plt.show()
```

```python
latest_date = df['date'].max()  # Get the most recent date
latest_population = df[df['date'] == latest_date]['population'].sum()
latest_total_vaccinations = df[df['date'] == latest_date]['total_vaccinations'].sum()

# Calculate the number of unvaccinated individuals
unvaccinated = latest_population - latest_total_vaccinations

# Create labels and sizes for the pie chart
labels = 'Vaccinated', 'Unvaccinated'
sizes = [latest_total_vaccinations, unvaccinated]
colors = ['skyblue', 'lightcoral']
explode = (0.1, 0)  # Explode the first slice (Vaccinated)

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.title(f'% Vaccinated vs. Unvaccinated of the Whole Population (as of {latest_date.strftime("%Y-%m-%d")})')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```
